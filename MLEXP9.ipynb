```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings as wr


wr.filterwarnings('ignore')


df = pd.read_csv("winequality-red.csv")
print(df.head())




print("Shape of the data:", df.shape)


df.info()


print(df.describe())


print("Columns in the dataset:", df.columns.tolist())


print("Missing values in each column:\n", df.isnull().sum())


print("Unique values in each column:\n", df.nunique())


quality_counts = df['quality'].value_counts()


plt.figure(figsize=(8, 6))
plt.bar(quality_counts.index, quality_counts, color='pink')
plt.title('Count Plot of Quality')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.xticks(quality_counts.index)
plt.show()




sns.set_palette("Pastel1")


plt.figure(figsize=(10, 6))
sns.pairplot(df)
plt.suptitle('Pair Plot for DataFrame', y=1.02)
plt.show()


plt.figure(figsize=(8, 6))
sns.boxplot(x='quality', y='alcohol', data=df)
plt.title('Box Plot between Alcohol and Quality')
plt.xlabel('Quality')
plt.ylabel('Alcohol')
plt.show()


plt.figure(figsize=(15, 10))
sns.heatmap(df.corr(), annot=True, fmt='.2f', cmap='Pastel2', linewidths=2)
plt.title('Correlation Heatmap')
plt.show()
```
